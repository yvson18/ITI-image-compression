```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def read_csv(directory, image):
    csv_files = [file for file in os.listdir(directory) if file.endswith('.csv')]
    csv_file = list(filter(lambda file: image in file, csv_files))[0]
    data = pd.read_csv(os.path.join(directory, csv_file))
    return data

# def calculate_inter_compressed_mean_sizes(dfs):
#     compressed_sizes = [df['compressed_size'] for df in dfs]
#     concatenated = pd.concat(compressed_sizes, axis=1)
#     print(concatenated)
#     row_means = concatenated.mean(axis=1)
#     return row_means

def get_image_size_in_kb(image):
    file_size_bytes = os.path.getsize(os.path.join("test_images", f"{image}.png"))
    file_size_kb = file_size_bytes / 1024
    return file_size_kb

def get_percentages(df, original_size):
    percentages = [100 * (compressed_size / original_size) for compressed_size in df['compressed_size']]
    return percentages

def plot_compression_curves(image):
    jpeg_data = read_csv("jpeg", image)
    bmshj2018_data = read_csv("bmshj2018-factorized-msssim", image)
    mbt2018_data = read_csv("mbt2018-mean-msssim", image)
    b2018_data = read_csv("b2018-gdn-192", image)

    print(jpeg_data)
    print(bmshj2018_data)
    print(mbt2018_data)
    print(b2018_data)

    original_size = get_image_size_in_kb(image)
    print(f"{original_size=}")
    # inter_compressed_mean_sizes = calculate_inter_compressed_mean_sizes([jpeg_data, bmshj2018_data, mbt2018_data])
    # size_ratios = [original_size / compressed_mean_size for compressed_mean_size in inter_compressed_mean_sizes]
    # print(inter_compressed_mean_sizes)
    # print(size_ratios)

    plt.plot(get_percentages(mbt2018_data, original_size), mbt2018_data['ssim'], marker='o', color='green', alpha=1, label='mbt2018', linestyle='-')
    plt.plot(get_percentages(bmshj2018_data, original_size), bmshj2018_data['ssim'], marker='o', color='red', alpha=1, label='bmshj2018', linestyle='-')
    plt.plot(get_percentages(b2018_data, original_size), b2018_data['ssim'], marker='o', color='purple', alpha=1, label='b2018', linestyle='-')
    plt.plot(get_percentages(jpeg_data, original_size), jpeg_data['ssim'], marker='o', color='blue', alpha=1, label='jpeg', linestyle='-')

    plt.title(f"Curvas de compressão para {image}")
    plt.ylabel('SSIM')
    plt.xlabel('Tamanho comprimido em relação ao original (%)')
    plt.legend(['mbt2018', 'bmshj2018', 'b2018', 'jpeg'], loc='lower right')
    # plt.tight_layout()
    plt.show()
```

```python
images = [os.path.splitext(file)[0] for file in os.listdir("test_images") if file.endswith('.png')]
for image in images:
    plot_compression_curves(image)
```
